## Take input from user and print table in that range. Suppose user input 5 and 10. Then print table of 5,6,7,8,9,10.

```python
tableValue = int(input("Enter value 1: "))
tableValue2 = int(input("Enter value 2: "))

print("\nFIRST OUTPUT:\n")

for i in range(tableValue, tableValue2):
    for j in range(1, 11):
        print (i,'x',j,'=',i*j)
    print()
    
print("\nSECOND OUTPUT:\n")
    
for x in range(1, 11):
    for y in range(tableValue, tableValue2):
        print(y, "x", x, "=", x*y, end= "\t")  
    print()
```
